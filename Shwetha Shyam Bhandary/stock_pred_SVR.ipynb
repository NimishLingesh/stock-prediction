```python
from scipy import stats
def featuring(data):
    data['ExpectedDividend'] = data['ExpectedDividend'].fillna(0)
    data["SupervisionFlag"] = data["SupervisionFlag"].astype(int)
    data['Target'] = data['Target'].fillna(0)
    
    cols = ['Open', 'High', 'Low', 'Close']
    data.loc[:,cols] = data.loc[:,cols].ffill()
    data.loc[:,cols] = data.loc[:,cols].bfill()

    data['Daily_Range'] = data['Close'] - data['Open']
    
    data['Open'] = stats.zscore(data['Open'])
    data['High'] = stats.zscore(data['High'])
    data['Low'] = stats.zscore(data['Low'])
    data['Close'] = stats.zscore(data['Close'])
    data['Volume'] = stats.zscore(data['Volume'])
    data['Daily_Range'] = stats.zscore(data['Daily_Range'])
    
    data = data.drop(['RowId', 'Date'], axis=1)
    
    return data
```

```python
data = featuring(stock_prices)
```

```python
df = data.sample(frac=0.15, replace=True, random_state=1)
df
```

```python
x = df.drop(['Target'], axis=1)
y = df['Target']
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size =0.2)
x_train.shape
```

```python
svr_ = svm.SVR()
svr_.fit(x_train, y_train)
```

```python
r_square = svr_.score(x_test, y_test)
```

R-square compares the models predictions to mean of the target/class label. Its value ranges from - infinity (poor model performance) to 1(better model). It will tell us how well our model has predicted the value. 

```python
print("Coefficent of determination for SVR:", r_square)
```

```python
#Mean Absolute Error(MAE)
from sklearn.metrics import mean_absolute_error

y_preds = svr_.predict(x_test)
mae = mean_absolute_error(y_test, y_preds)
mae
```

```python
#Mean Squared Error(MSE)
from sklearn.metrics import mean_squared_error
y_preds = svr_.predict(x_test)
mse = mean_squared_error(y_test, y_preds)
mse
```
